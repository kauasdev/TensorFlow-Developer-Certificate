### 3.Create two tensors containing random values between 0 and 1 with shape `[5, 300]`.

```python
tensor1 = tf.random.uniform(shape=[5, 300])
tensor2 = tf.random.uniform(shape=[5, 300])

tensor1, tensor2
```

### 4.Multiply the two tensors you created in 3 using matrix multiplication.

```python
tensor1 * tensor2
```

### 5.Multiply the two tensors you created in 3 using dot product.

```python
tf.tensordot(tensor1, tf.transpose(tensor2), axes=0)
```

```python
tf.tensordot(tensor1, tf.transpose(tensor2), axes=1)
```

```python
tf.tensordot(tf.transpose(tensor1), tensor2, axes=0)
```

```python
tf.tensordot(tf.transpose(tensor1), tensor2, axes=1)
```
